#### We use euclidean_distances() method to find distance between rows. So we can understand how  ideologically similar they are

```python
from sklearn.metrics.pairwise import euclidean_distances

distance_0_and_1 = euclidean_distances(votes.iloc[0,3:].values.reshape(1, -1), votes.iloc[1,3:].values.reshape(1, -1))

distance_0_and_2 = euclidean_distances(votes.iloc[0,3:].values.reshape(1, -1), votes.iloc[2,3:].values.reshape(1, -1))

print("distance between first and second row :",distance_0_and_1)

print("distance between first and third row :",distance_0_and_2)
```

#### Use KMeans class to find clusters. 

```python
import pandas as pd
from sklearn.cluster import KMeans

kmeans_model = KMeans(n_clusters=2, random_state=1)
senator_distances = kmeans_model.fit_transform(votes.iloc[:,3:])

senator_distances
```

First column is euclidean distance between senator and first cluster.Second column is distance between senator and second cluster

#### Extract the cluster labels for each Senator from kmeans_model using kmeans_model.labels_, then we can make a table comparing these labels to votes["party"] with crosstab(). This will show us if the clusters tend to break down along party lines or not

```python
labels = kmeans_model.labels_
pd.crosstab(labels,votes["party"])
```

#### Code will select all Independents in the first cluster

```python
independents = votes[(labels == 0) & (votes["party"] == "I")]
independents
```

####  All senators who were assigned to the second cluster that were Democrats

```python
democratic_outliers = votes[(labels == 1)&(votes["party"] == "D")]
democratic_outliers
```

####  Make a scatterplot that shows the position of each Senator. This works because the distances are relative to the cluster centers.

```python
import matplotlib.pyplot as plt

x = senator_distances[:,0]
y = senator_distances[:,1]


plt.scatter(x,y,c=labels)
plt.show()
```

#### Compute an extremism rating by cubing every value in senator_distances, then finding the sum across each row

```python
import numpy as np
extremism = np.sum(senator_distances**3,axis=1)
votes["extremism"] = extremism
votes.sort_values("extremism", inplace=True, ascending=False)
votes.head(10)
```
